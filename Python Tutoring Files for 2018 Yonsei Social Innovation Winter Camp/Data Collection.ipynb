사람이 쓰거나 기계로 인쇄한 문자의 영상을 이미지 스캐너로 획득하여 기계가 읽을 수 있는 문자로 변환하는 기술

```python
#pip3 install pytesseract
#pip3 install gTTs
from PIL import Image
import pytesseract
import argparse
import cv2
import os
from gtts import gTTS
import unicodedata

# Path of image that you want to process OCR
image_path = '/Users/angieryu2202/Desktop/dust.jpg'

# Using OpenCv to read the image
image = cv2.imread(image_path)

# Processing the image into black & white
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# apply thresholding to preprocess the
# image

gray = cv2.threshold(gray, 0, 255,
        cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

# remove noise
gray = cv2.medianBlur(gray, 3)

# write the grayscale image to disk as a temporary file so we can
# apply OCR to it
filename = "{}.png".format(os.getpid())
cv2.imwrite(filename, gray)

# load the image as a PIL/Pillow image, apply OCR, and then delete
# the temporary file
text = pytesseract.image_to_string(Image.open(filename))
os.remove(filename)


text = unicodedata.normalize('NFKD', text).encode('ascii','ignore')
print(text)
```

#### Example 1 :  Yonhapnews Data Crawling


- Gather more than 100 news article data from Yonhap News.

- The data should include  "article contents", "reporter", "title", "date", and "company name"

```python
from bs4 import BeautifulSoup
import lxml
import urllib.request
```

```python
url_first = "http://www.yonhapnews.co.kr/culture/090100000"
url_last = ".html"

# Create an empty list to store article urls
news_article_urls = []

# Iterating through page 1 to 5
for url_middle in range(1,6):
    # Creating url
    url = url_first + str(url_middle) + url_last
    
    # Retrieving raw html
    raw_html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(raw_html, 'lxml')
    
    # Each link of news articles are stored in:
    ## <strong class="news-tl"><a href="HERE!"> 
    #### Get all <strong class="news-tl"> tags
    content = soup.find_all("strong",{"class":"news-tl"})
    
    #### From each <strong class="news-tl">, only retrieve <a href>
    for title_link in content:
        one_link = title_link.find('a')["href"]
        news_article_urls.append(one_link)
    
```

```python
# Let's see how many articles we got
print(len(news_article_urls))
```

```python
# Retrieve data from each articles

# Process indicator
process = "Crawling article number: "

# Making empty dictionary
newsdata = {}

for x in range(0,125):
    
    # creating article dictionary inside the dictionary
    newsdata[x+1] = {}
    
    url = news_article_urls[x]
    raw_html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(raw_html, 'lxml')
    
    # Newspaper company
    newscomp = soup.find("meta",{"property":"og:site_name"})["content"]
    
    # Title
    title = soup.find("meta",{"property":"og:title"})["content"]
    
    # Reporter
    reporter = soup.find("meta",{"itemprop":"publisher"})["content"]
    
    # Date
    date = soup.find("meta",{"property":"article:published_time"})["content"].split("T")[0]
    
    # Content
    content = soup.find("div",{"class":"article"}).text.split("@")[0].strip().replace("\t","").replace("\r","").replace("\n","")

    
    # Add all these info to the list
    newsdata[x+1]["newscomp"] = newscomp
    newsdata[x+1]["titles"] = title
    newsdata[x+1]["reporters"] = reporter
    newsdata[x+1]["dates"] = date
    newsdata[x+1]["contents"] = content
    
    # print process to know how your crawler is doing
    print(process + str(x+1))   

```

```python
import pandas as pd

df = pd.DataFrame(newsdata)
df = df.transpose()

df.head()
```

```python
# Saving the data into file

df.to_csv("/Users/angieryu2202/Desktop/news.tsv", encoding="CP949", sep="\t")
```
